## libcudnn 9.1.0への対応

NVIDIA Cosmosでは、libcudnn 9.1.0が必要だが、`apt-get install libcudnn9-cuda-12`では、9.6.0がインストールされるので、臨時の対策でシンボリックを貼ることで対処(おすすめできない方法）



````python
!ln -sf /usr/lib/x86_64-linux-gnu/libcudnn_ops.so.9 /usr/lib/x86_64-linux-gnu/libcudnn_ops.so.9.1.0
````

## Cosmos-1.0-Autoregressive-4B を実行

https://github.com/NVIDIA/Cosmos/blob/main/cosmos1/models/autoregressive/README.md

````python
!CUDA_VISIBLE_DEVICES=0 PYTHONPATH=$(pwd) python cosmos1/models/autoregressive/inference/base.py \
    --input_type=video \
    --input_image_or_video_path=cosmos1/models/autoregressive/assets/v1p0/input.mp4 \
    --video_save_name=Cosmos-1.0-Autoregressive-4B \
    --ar_model_dir=Cosmos-1.0-Autoregressive-4B \
    --top_p=0.8 \
    --temperature=1.0
````

## 自分の画像から生成

`input_image_or_video_path` にColabにアップした自分の画像を指定。run.mp4をアップロードした場合は、/content/run.mp4 のように指定する


````python
!CUDA_VISIBLE_DEVICES=0 PYTHONPATH=$(pwd) python cosmos1/models/autoregressive/inference/base.py \
    --input_type=video \
    --input_image_or_video_path=/content/run.mp4 \
    --video_save_name=Cosmos-1.0-Autoregressive-4B \
    --ar_model_dir=Cosmos-1.0-Autoregressive-4B \
    --top_p=0.8 \
    --temperature=1.0
````
